```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
esg = pd.read_csv("danl_210_PACKERT_THEODORE_ESG.csv")
stock = pd.read_csv("danl_210_PACKERT_THEODORE_stock.csv")

stock['Date'] = pd.to_datetime(stock['Date'])
esg.head()
```

```python
print("📊 ESG Summary:")
print(esg.describe(include='all'))

print("\n📈 Stock Summary:")
print(stock.describe())
```

```python
# Fix: Ensure numeric types for plotting
stock[['Open', 'High', 'Low', 'Close']] = stock[['Open', 'High', 'Low', 'Close']].apply(pd.to_numeric, errors='coerce')

# Re-plot with corrected values
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sample = stock[stock['Symbol'].isin(['AAPL', 'MSFT', 'NVDA'])]
sns.lineplot(data=sample, x='Date', y='Close', hue='Symbol')
plt.title("Stock Closing Prices (2024–Q1 2025)")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```
